```python
import sys, json, time, random

# 项目路径
SIMWORLD_DIR      = r"D:\LLMDelivery-LJ\SimWorld"
LLM_DELIVERY_DIR  = r"D:\LLMDelivery-LJ\LLM-Delivery"

sys.path.insert(0, SIMWORLD_DIR)
sys.path.insert(0, LLM_DELIVERY_DIR)

from Base.Map import Map
from Base.Order import OrderManager
from Base.DeliveryMan import DeliveryMan, TransportMode

# 数据文件
ROADS_JSON        = r"D:\LLMDelivery-LJ\Test_Data\test\roads.json"
WORLD_JSON        = r"D:\LLMDelivery-LJ\Test_Data\test\progen_world_enriched.json"

# ----------------- 小工具 -----------------
def _load_world_nodes(path: str):
    with open(path, "r", encoding="utf-8") as f:
        obj = json.load(f)
    return obj.get("nodes", [])

m = Map()
m.import_roads(ROADS_JSON)
m.import_pois(WORLD_JSON)

# 准备订单池
nodes = _load_world_nodes(WORLD_JSON)
om = OrderManager(capacity=10)
om.fill_pool(m, nodes)

# 创建送餐员
dm = DeliveryMan(city_map=m, world_nodes=nodes, x=0.0, y=0.0, mode=TransportMode.BIKE, energy_max=120.0)

# 看地图信息（可选）
info = dm.agent_info()

print(info["text"])

# 接第一单（接完自动补齐）
order = om.list_orders()[0]
dm.accept_order(order, order_manager=om)

print(dm.to_text())
```
